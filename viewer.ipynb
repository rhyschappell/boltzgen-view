# BoltzGen Design Viewer

Interactive parallel-coordinates viewer for BoltzGen protein design outputs.

```python
design_dir = "/path/to/intermediate_designs_inverse_folded"

rmsd_threshold = 2.5

# Additional filters beyond the defaults, e.g.:
extra_filters = [{"feature": "design_to_target_iptm", "lower_is_better": False, "threshold": 0.5}]

max_display = 5000
```

```python

import copy

from boltzgen_view.loader import load_metrics
from boltzgen_view.filters import apply_filters, DEFAULT_FILTERS, print_summary

df = load_metrics(design_dir)
print(f"Loaded {len(df):,} designs from {design_dir}")
print(f"Columns: {list(df.columns)}\n")

filters = copy.deepcopy(DEFAULT_FILTERS)
for f in filters:
    if f["feature"] in ("filter_rmsd", "filter_rmsd_design"):
        f["threshold"] = rmsd_threshold

df_filtered, summary = apply_filters(df, filters=filters, extra_filters=extra_filters, max_display=max_display)
print_summary(summary)
```

```python
from boltzgen_view.viewer import ParallelCoordsViewer

viewer = ParallelCoordsViewer(
    df_filtered,
    design_dir=design_dir,
    max_display=max_display,
)
viewer.show()
```

```python
selected = viewer.get_selected()
print(f"{len(selected):,} designs selected")

show_cols = [c for c in ["id", "design_to_target_iptm", "design_ptm",
                          "min_design_to_target_pae", "filter_rmsd"]
             if c in selected.columns]
display(selected[show_cols])
```
